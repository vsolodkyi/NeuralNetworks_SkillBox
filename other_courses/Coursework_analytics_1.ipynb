```python
print('Количество уроков в каждом модуле на каждом курсе:', '\n')
content_grouped = main.groupby(['course_title', 'module_number']).max()['lesson_number'].to_frame()

# можно было отобразить длинный DataFrame (229 элементов) с помощью опции: pd.set_option('display.max_rows', 300)
# но в этом случае при пролистывании теряется шапка таблицы, и особого удобства это не добавляет
# Поэтому сделал перебор по каждому курсу

for i in list(content_grouped.index.get_level_values(0).unique()):
    #pass
    print(f'Курс: {i}\n\n{content_grouped.loc[i,]}\n\n')

# подсчет медианного количества уроков на модуль в каждом курсе
print('Медианное количество уроков в модуле на каждом курсе: ')
content_grouped.reset_index().groupby('course_title')['lesson_number'].agg(median_lessons='median')
```

```python
print('Количество учеников на каждом курсе:')

# датафрейм с общим количеством уроков в каджом курсе
course_sum_df = content_grouped.groupby('course_title')['lesson_number'].sum().reset_index() 

# датафрейм с количеством законченных уроков по каждому прогрессу и общее количество модулей в этом же курсе
temp3 = main[main['status'] == 'done'].groupby(['progress_id', 'course_title', 'student_id']).count()['status'].reset_index()\
    .merge(right=course_sum_df, how='left', on='course_title')

# список выпускников соответствующих курсов, т.е. те прогрессы, на которых N_уроков = N_законченных уроков
graduated_df = temp3[temp3.status == temp3.lesson_number][['student_id', 'course_title']]

# объединяем общее количество студентов по каждому курсу с общим количеством выпускников этого курса
students_by_course_df = main.groupby('course_title')['student_id'].nunique().to_frame()\
                            .merge(right=graduated_df.groupby('course_title').count(), on='course_title', how='left')\
                            .fillna(0)
# наводим косметику
students_by_course_df.columns = ['student_total', 'student_graduated']

# считаем колиечство студентов, которые еще не закончили обучение
students_by_course_df['student_active'] = students_by_course_df['student_total'] - students_by_course_df['student_graduated']

students_by_course_df = students_by_course_df.astype({'student_graduated' : 'int32', 'student_active' : 'int32'})
students_by_course_df.sort_values('student_total', ascending=False)
```

```python
# Переводим к анализу дат рождения студентов
# pd.read_csv не смог спарсить даты рождения студентов. Поэтому в начале ищем те строки, которые парсятся неверно
wrong_records = []
for i in students_df.index:
    if type(pd.to_datetime(students_df.loc[i, 'student_birthday'], errors='ignore')) == str:
        wrong_records.append(students_df.loc[i, 'student_id'])
print('Студенты с явно неправильной датой роджения:')
students_df[students_df['student_id'].isin(wrong_records)]
```

```python
# заменяем все даты выше на NaT и переводим даты в DateTime тип
students_df['student_birthday'] = pd.to_datetime(students_df['student_birthday'], errors='coerce')

# функция считает возраст
def calculate_age(date):
    today = pd.to_datetime('today')
    return today.year - date.year - ((today.month, today.day) < (date.month, date.day))

students_df['age'] = students_df['student_birthday'].apply(calculate_age)

# можем заметить, что достаточно много студентов в базе с неправильным возрастом ()
print('Студентов с подозрительным возрастом (>90 или <10): ',\
      students_df[(students_df['age'] > 90) | (students_df['age'] < 10)].shape[0])
```

```python
# задача №1: создать словарь first_HW = {course_id: (module, lesson), ...}, 
# где (module, lesson) - индекс первого домашнего задания. Ограничиваемся content_df вместо main для быстродействия
home_work_df =  content_df[content_df['is_homework'] == True]

# проходим по всем урокам с флагом is_homework и записываем наименьшую пару (module_number, lesson_number)
first_HW = dict()
for i in home_work_df.index:
    course = home_work_df.loc[i, 'course_id']
    lesson_index = (home_work_df.loc[i, 'module_number'], home_work_df.loc[i, 'lesson_number'])
    if first_HW.get(course):
        if lesson_index < first_HW[course]:
            first_HW[course] = lesson_index
    else:
        first_HW[course] = lesson_index

# задача №2: содаем нулевой датафрейм: индекс - курсы, столбцы - месяца с 2016-03 по 2019-07
rows = list(first_HW.keys())
columns = [i.strftime('%Y-%m') for i in pd.date_range(start='2016-03-01', end='2019-08-01', freq='m')]
student_increment_df = pd.DataFrame(0, index=rows, columns=columns)
student_increment_df.index.name = 'course_id'

# задача 3: считаем первые попытки выполнения ДЗ.
# предположим, что студент может пытаться выполнить любое (даже первое) домашнее задание несколько раз,
# сдедовательно мы будем считать только первые попытки выполнения ДЗ:
student_start_date_df = main[main['is_homework'] == True]\
                        .groupby(['course_title', 'course_id', 'student_id', 'module_number', 'lesson_number'])['start_date']\
                        .min()\
                        .reset_index()

# задача №4: заполняем датафрейм данными о количестве новых студентов в каждом месяце
for i in student_start_date_df.index:
    course = student_start_date_df.loc[i, 'course_id']
    if (student_start_date_df.loc[i, 'module_number'] == first_HW[course][0]) and \
            (student_start_date_df.loc[i, 'lesson_number'] == first_HW[course][1]):
        student_increment_df.loc[course, student_start_date_df.loc[i, 'start_date'].strftime('%Y-%m')] += 1 
```

```python
# задача №1: создать словарь module_HWs = {course_id: {module1 : Module(first lesson, last lesson), module2: ..}, ...} 
# где класс Module хранит номер урока первой и последней домашней работы.
class Module:
    def __init__(self, lesson):
        self.first_hw = lesson
        self.last_hw = lesson
    
    def __str__(self):
        return f'first_hw:{self.first_hw}, last_hw:{self.last_hw}'
    
    def __repr__(self):
        return f'first_hw:{self.first_hw}, last_hw:{self.last_hw}'

home_work_df =  content_df[content_df['is_homework'] == True]

# проходим по всем урокам с флагом is_homework и составляем словарь курс: модуль: номер первого и последнего ДЗ.
module_HWs = dict()
for i in home_work_df.index:
    course = home_work_df.loc[i, 'course_id']
    module = home_work_df.loc[i, 'module_number']
    lesson = home_work_df.loc[i, 'lesson_number']
    
    if module_HWs.get(course):  # if course exists in dict
        if module_HWs[course].get(module):# if module exists in dict
            if lesson < module_HWs[course][module].first_hw:
                module_HWs[course][module].first_hw = lesson # update first HW lesson
            if lesson > module_HWs[course][module].last_hw:
                module_HWs[course][module].last_hw = lesson # update last HW lesson
        else: # course exists, module doesn't
            module_HWs[course][module] = Module(lesson) 
    else: # course doesn't exist
        module_HWs[course] = {module: Module(lesson)}
        
# теперь у нас есть словарь с индексом первого и последнего урока module_HWs[course][module].first_hw or .last_hw

```

```python
# создадим пустой DataFrame для дополнения в него информацию по каждому модулу и студенту, успешно прошедшему модуль
module_duration_df = pd.DataFrame(columns=['course_id', 
                                           'course_title', 
                                           'module_number', 
                                           'student_id', 
                                           'start_date', 
                                           'finish_date', 
                                           'duration'])
module_duration_df = module_duration_df.astype({'duration' : 'int32', 
                                                'module_number' : 'int32'})

# датафрейм с неоконченными модулями
unfinished_modules_df = pd.DataFrame(columns=['course_id', 
                                           'course_title', 
                                           'module_number', 
                                           'student_id', 
                                           'start_date', 
                                           'finish_date', 
                                           'duration'])
unfinished_modules_df = module_duration_df.astype({'duration' : 'int32', 
                                                'module_number' : 'int32'})

today_date = pd.to_datetime('now', utc=True) # считаем статистику по длительности модуля для незаконценных модулей

# Очень долгий цикл
for name, df in main_filtered[main_filtered['status'] == 'done']\
                    .groupby(['course_id', 'course_title', 'module_number', 'student_id']):
    # паспаковываем итератор
    course, course_title, module, student = name
    # ищем самый ранний старт первого ДЗ
    start = df[df['lesson_number'] == module_HWs[course][module].first_hw]['start_date'].min()
    # ищем самое позднее окончание последнего ДЗ модуля
    finish = df[df['lesson_number'] == module_HWs[course][module].last_hw]['finish_date'].max()
    
    # есть случай когда первая домашняя работа так и осталась незавершенной со статусом fail. Делаем обработку..
    # Игнорируем такие строки, но запишем в отдельный дата фрейм для анализа
    if pd.isnull(start):
        duration = np.nan
        unfinished_modules_df = unfinished_modules_df.append({'course_id' : course,
                                                    'course_title' : course_title, 
                                                    'module_number' : module, 
                                                    'student_id' : student, 
                                                    'start_date' : start, 
                                                    'finish_date' : finish, 
                                                    'duration': duration}, ignore_index=True)
        continue
    # finish is null в случае, если студент еще не закончил текущий модуль = не выполнил последнее ДЗ модуля
    # Игнорируем такие строки, но запишем в отдельный дата фрейм для анализа
    if pd.isnull(finish):
        duration = int((today_date - start).total_seconds())
        unfinished_modules_df = unfinished_modules_df.append({'course_id' : course,
                                                    'course_title' : course_title, 
                                                    'module_number' : module, 
                                                    'student_id' : student, 
                                                    'start_date' : start, 
                                                    'finish_date' : finish, 
                                                    'duration': duration}, ignore_index=True)
        continue
        
    duration = int((finish-start).total_seconds())
    module_duration_df = module_duration_df.append({'course_id' : course,
                                                    'course_title' : course_title, 
                                                    'module_number' : module, 
                                                    'student_id' : student, 
                                                    'start_date' : start, 
                                                    'finish_date' : finish, 
                                                    'duration': duration}, ignore_index=True)
```

### 4.1. Расчет конверсии.

```python
# формируем новый датафрейм convertion_df наполняя его информацией по каждому модулю: сколько начали, сколько студентов окончили

conversion_df = pd.DataFrame(columns=['course_id', 'module_number', 'students_started', 'students_finished'])

# module_HWs - содержит данные по номеру первого и последнего ДЗ для каждого модуля каждого курса
for course in module_HWs.keys():
    for module in module_HWs[course].keys():
        # фильтруем ДЗ по текущему итерируемому курсу-модулю
        mask = (main_filtered['course_id'] == course) & (main_filtered['module_number'] == module)
        # считаем количество уникальных студетнов, начавших первое ДЗ
        started = main_filtered[(main_filtered['lesson_number'] == module_HWs[course][module].first_hw) & mask]['student_id']\
                    .nunique()
        # считаем количество уникальных студетнов, закончивших последнее ДЗ, исключая незаконченные ДЗ
        finished = main_filtered[(main_filtered['lesson_number'] == module_HWs[course][module].last_hw) & mask\
                                & (~pd.isnull(main_filtered['finish_date']))]['student_id'].nunique()
        # заполняем датасет
        conversion_df = conversion_df.append({'course_id' : course,
                                              'module_number' : module,
                                              'students_started' : started,
                                              'students_finished' : finished}, ignore_index=True)
```

```python
# добавляем колонку course_title, удаляем course_id, сортируем по курс-модуль
conversion_df = conversion_df.merge(courses_df[['course_id', 'course_title']], on='course_id', how='left')\
                    .drop('course_id', axis=1).sort_values(['course_title', 'module_number'])\
                    .set_index(['course_title', 'module_number'])
```

```python
conversion_df = conversion_df.reset_index() # сбрасываем индекс для удобства в следующем блоке кода
```

```python
conversion_percent = [] # формируем список с расчитанной конверсией для добавления с датафрейм в конце
# датафрем отсотрирован, поэтому исключаем проверку на то, что модули идут не по возрастанию
for course, data in conversion_df.groupby('course_title'):
    next_course = 0 # индикатор, который используем для определения самой первой итерации цикла
    finished = 0 # сколько студентов закончило предыдущий модуль
    # итератор i совпадает с индексом в conversion_df, и перебор значений уже идет по порядку, проверял с print(i) в цикле
    # поэтому сам индекс в списке не храним
    for i in data.index:
        # первый модуль с ДЗ обрабатывается отдельно
        if next_course == 0:
            conversion_percent.append(100)
            next_course = 1
            finished = data.loc[i, 'students_finished']
            continue
        started = data.loc[i, 'students_started']
        # исключаем деление на ноль, заменяя значение конверсии на 0 (только один подобный случай на курсе Анимации, модуль 13)
        if finished == 0:
            # print(data.loc[i]) # строка для debugging алгоритма и оценки количества таких случаев
            conversion_percent.append(0)
            finished = data.loc[i, 'students_finished']
            continue    
        conversion_percent.append(started / finished * 100)
        finished = data.loc[i, 'students_finished']
        
conversion_df['conversion_percent'] = conversion_percent
conversion_df
```

## 5. Метрика успеваемости 

&nbsp;&nbsp;&nbsp;&nbsp;*(необязательное задание)*

Иногда студенты берут курсы, которые оказываются для них неподъемными. Это может быть как по причинам недостаточной изначальной подготовки, так и по причинам, связанным с низкой мотивацией студента. Для улучшения качества контента полезно выявить причину. На основании имеющихся данных придумайте метрику успеваемости студента. **Обоснуйте ее.** Выявите таких студентов на каждом курсе, чтобы предоставить продюсерам список проблемных студентов. 

**Решение**


Всё познается в сравнении. Важно текущее состояние мотивации. Оцениваем по последним 2-3 домашним заданиям.Учитываем скорости выполнения текущей и прошлых ДЗ, в стравнении с медианным значением. Используем взвешенное среднее последних 3-х заданий. 


Rating = 0.5*[(duration_median_i   - duration_i)  / (duration_median_i   + duration_i)]   + <br>
&nbsp;&nbsp; 0.3*[(duration_median_i-1 - duration_i-1)/ (duration_median_i-1 + duration_i-1)] + <br>
&nbsp;&nbsp; 0.2*[(duration_median_i-2 - duration_i-2)/ (duration_median_i-2 + duration_i-2)]

где i - последний модуль с ДЗ, который студент начал (если еще не закончил, то duration считается до текущего числа)

Причина весовых коэфициентов, учет влияния мотивации,

Данный рейтинг изменяется в диапазоне от -1 до 1. Для удобства перенормируем его от 0 до 1, где 
- 0 - студент выполняет домашние задания значительно медленнее коллег, 
- 0.5 - абсолютный середнячок, выполняет ДЗ так же быстро, как и медианное значение
- 1 - студент выполняет домашние задания значительно быстрее медианного значения.


Далее мы можем выбрать самых отстающих (по уровню и/или мотивации) выбирая студентов с рейтингом, например менее 0.1. Рейтинг 0.1 соответствует студенту, который выполняет ДЗ в 9 раз медленнее медианного значения.


```python
# расчет отношения скорости выполнения ДЗ к медианному значению по рейтингу
rating = 0.1
ratio = (1 - rating) / rating
print(f'Рейтинг {rating} сооответсвует отношению {ratio}:1')
```

```python
# расчет рейтинга по отношению скорости выполнения ДЗ к медианному значению
ratio = 9
rating = 1/(1+ratio)
print(f'Отношению {ratio}:1 сооответсвует рейтингу {rating}')
```

```python
student_rating = pd.DataFrame(columns=['student_id', 'course_title', 'last_module', 'rating'])
# предусматриваем 3 набора весовых коэффициентов, в зависимости от количества последних начатых домашних работ: 1, 2 или 3
weights = [[1], [0.7, 0.3], [0.5, 0.3, 0.2]]

#если ДЗ не завершенно, считаем продолжительность выполнения ДЗ на сегоднешнее число
today_date = pd.to_datetime('now', utc=True)

# расчет длительности выполнения ДЗ
def calculate_duration(start_date, finish_date):
    if pd.isnull(finish_date):
        finish_date = today_date
    return int((finish_date - start_date).total_seconds())

# выгрузим датафрейм с медианными значениями длительности выполнения ДЗ
median_duration = module_duration_df.groupby(['course_title', 'module_number'])['duration_days']\
                                    .median() # Series
# выгружаем список студентов
students_list = main_filtered['student_id'].unique()

# сортируем датафрейм с домашними заданиями по дате начала, чтобы позже выбирать последние три ДЗ
main_sorted = main_filtered.sort_values('start_date')

# для каждого студента считаем рейтинг
j=0
for student in students_list:
    
    data = main_sorted[main_sorted['student_id'] == student].iloc[-3:, :].copy()
    # для всех 3 строк считаем длительность выполнения задания
    data['duration'] = data.apply(lambda row: calculate_duration(row['start_date'], row['finish_date']), axis=1)
    # last course/module внесем в датафрейм
    last_course_title = data.iloc[-1]['course_title']
    last_module = data.iloc[-1]['module_number']
    
    # перебираем 3 (или 2, или 1) строки с последними ДЗ и с весовыми коэффициентами складываем
    rating = 0
    for i in range(data.index.size):
        course = data.iloc[i]['course_title'] # course_title
        module = data.iloc[i]['module_number'] # module_number
        # редкое исключение, когда есть модуль, который никто еще не закончил, в этом случае за этот модуль все 
        # студенты, кто начал этот модуль, получат рейтинг 0.5
        try:
            median = median_duration[course, module] * 24 * 3600 # переводим дни в секунды
        except:
            median = data.iloc[i]['duration']
            j += 1
            
        rating += weights[data.index.size-1][i] *\
                  ((median - data.iloc[i]['duration']) / (median + data.iloc[i]['duration']) + 1) / 2
    # дописываем с датафрейм
    student_rating= student_rating.append({'student_id' : student, 
                                           'course_title' : last_course_title, 
                                           'last_module' : last_module, 
                                           'rating' : rating}, ignore_index=True)
```

```python
student_rating = student_rating.astype({'last_module' : 'int32'})
```

```python
# данная таблица приводит анти-рейтинг курсов, где больше всего проблемных студентов, возможно, дело не в студентах...
print('Количество студентов с рейтингом менее 0.1 для каждого курса и статистика по последнему запущенному модулу')
student_rating[student_rating['rating'] < 0.1].sort_values('rating').groupby('course_title')\
    .agg({'student_id':'count', 'last_module': ['min', 'max', 'median']})
```

Из таблицы выше видим, что многие студенты скорее всего бросили курс на 1-4 модуле, при этом с самого начала изрядно отставая от прогресса успешных учеников. Возможно, имеет смысл ввести пояснительную секцию в описание курса с рекомендациями по предварительной подготовке до того, как записываться на курс. Несмотря на то, это это будет иметь прямой негативный экономический эффект, всё же это может повысить количество выпустников, также может повлиять на корпоративных заказчиков, стимулируя их с отправке сотрудников на курсы повышения квалификации, наблюдая высокий процент выпускников - что в кнечном итоге может оказаться более выигрышным.
