Проанализируем ответы на Q1. Отсслеживаем случаи, когда в ответе:
- число > 10
- отсутствует ответ
- отсутствует число либо чисел больше одного

```python
counter = dict()
for answer in df.Q1.unique():
    if pd.isna(answer):
        counter[answer] = len(df[df.Q1 == answer])
        continue
    options = re.findall(r'[0-9]+', answer)
    if len(options) == 0:
        counter[answer] = len(df[df.Q1 == answer])
    elif len(options) > 1:
        counter[answer] = len(df[df.Q1 == answer])
    elif int(options[0]) > 10: # and len == 1
        counter[answer] = len(df[df.Q1 == answer])
print('number of irregular answers:', sum([v for k, v in counter.items()]))
print(
    'Frequent irregular answers:\n' +\
    '\n'.join([f'  answer: {k:10} frequency: {v}' for k,v in counter.items() if v > 1])
     )
```

Так все некорректные ответы встречаются не более 2 раз и всего их 38 (1.2%), то все их выкидываем (заменяем на None), за исключением 'Очень  хорошо ... 555', который однозначно можно разметить как максимальный бал, в отличии от 11, 15, 19 - что могут являться опечатками. Также, если даны две оценки, отличающиеся не более чем на 2, то результат округляем.

```python
def parse_Q1(s):
    """
    Parse Q1 answer with heuristic rules.
    s is a string with user's answer, where expected value is
    a whole number, but could contain several numbers or nothing.
    """
    # skip nan values
    if pd.isna(s):
        return None
    
    # parse numbers
    options = re.findall(r'[0-9]+', s)
    
    # apply rules
    # no numbers in answer
    if len(options) == 0:
        return None
    
    # more than two numbers in answer
    if len(options) > 1:
        nums = [int(i) for i in options]
        if max(nums) - min(nums) <= 2:
            return np.mean(nums)
        else:
            return None
    # one single number
    else:
        rating = int(options[0])
        # case: "Очень  хорошо. Обслуживания  я довольно. Спасибо вам.555"
        rating = rating if rating != 555 else 10
        # as answer is in range [1, 10], replace 0 with 1
        rating = rating if rating != 0 else 1 
        if rating > 10:
            return None
        return rating
```

```python
df['Q1_answer'] = df.Q1.apply(parse_Q1)
```

Проанализируем ответы на второй вопрос. Возможные варианты ответа: числа от 1 до 7, перечисленные через запятую. Среди возможных значений есть двухзначные числа и 0, которые игнорируем.

```python
def parse_Q2(s, target_question=None):
    """
    Parse Q2 answer with heuristic rules.
    Expecting a string with comma separated numbers
    in range [1, 7].
    return 1 if target_question is a string, else 0
    """
    # skip nan values
    if pd.isna(s):
        return [0] * 7
    
    # parse numbers
    options = [int(i) for i in re.findall(r'[0-9]+', s)]
    return [1 if i+1 in options else 0 for i in range(7)]
```

```python
df['Q2_answers'] = df.Q2.apply(parse_Q2)
```

На основании выводов о частотах отдельных опций в вопросе Q2 можно предположить, что страт должно быть не много.

```python
from sklearn.cluster import KMeans
from sklearn.metrics.pairwise import euclidean_distances


metrics = []
MAX_CLUSTERS = 10
X = df.loc[:, 'Q2_1':'Q2_6'].values

for cluster_num in range(1, MAX_CLUSTERS):
    model = KMeans(n_clusters=cluster_num, random_state=42).fit(X)
    centroids, labels = model.cluster_centers_, model.labels_
    metric = 0
    for centroid_label in range(cluster_num):
        metric += euclidean_distances(
            X[labels==centroid_label],
            centroids[centroid_label,:].reshape(1,-1)
        ).sum()
    metrics.append(metric)

# use the elbow method to determine the optimal number of clusters
D = []
for i in range(0, len(metrics)-1):
    d = abs((metrics[i+1] - metrics[i]) / (metrics[i] - metrics[i-1]))
    D.append(d)
print('optimal number of clasters (in descending order):', np.argsort(D) + 1)

plt.plot(range(1, len(metrics)+1), metrics)
plt.title('Sum of distances of $X_i$ to its cluster center')
plt.xlabel('number of clusters'); plt.ylabel('total distance')
plt.show()
```

```python
def distribution_alalysis(data, params, condition, category_names, 
                          title=None, figsize=(15,30)):
    """
    Plot distribution analysis of params in data, splitting data
    into two groups based on condition.
    Inputs:
        data - dataframe with data
        params - keys of columns in data to be plotted and analized
        condition - string expression resulting in bool value to split data 
            into 2 groups, which is compatible with pd.DataFrame.eval(),
            i.e. "Quality=='below_avg'" (key-condition-value)
        title - general title of the plot, None values gives default title
        category_names - list of two strings: first corresponds to condition
            (group 1), second - to the rest of data (group 2)
        figsize - plot figure size as used for matplotlib.pyplot
    """
    n_rows = len(params)
    mask = data.eval(condition)
    labels = np.where(mask, category_names[0], category_names[1]).tolist()
    if title is None:
        title = 'Distribution analysis of parameters in two groups'
    
    plt.figure(figsize=figsize)
    for i, param in enumerate(params):
        ax = plt.subplot(n_rows, 2, 2*i+1)
        # as long as cardinalities of 2 groups are not equal, normalize traffic with weights
        if param == 'Total Traffic(MB)':
            weight = sum(mask) / sum(~mask)
            weights = mask * 1 + (~mask) * weight
            bins = 16
        else:
            weights = None
            bins = 'auto'
        ax.set_title(f'histogram of {param}')
        sns.histplot(data=data, x=param, hue=labels, ax=ax, weights=weights, bins=bins)
        # scale y axis for technical parameters
        if param != 'Total Traffic(MB)':
            ax.set_yscale('log')

        ax = plt.subplot(n_rows, 2, 2*i+2)
        sns.ecdfplot(data=data, x=param, hue=labels, ax=ax)
        ax.set_title(f'CDP of {param}')

    plt.tight_layout(pad=2.0)
    plt.suptitle(title, y=1)
    plt.show()
```

#### 1. Статистический тест

Так как замеры технических параметров разными пользователями можно рассматривать как различные измерения одного и того же объекта (определенная телекоммуникационная вышка например), то предполагая что все абоненты живут в одном маленьком городе, то мы воспользовались парным t-test'ом (ttest_rel), но если речь о Москве или другом крупном городе, а также принимая во внимание мобильность связи (один и тот же абонент "собирает" данные о тех.показателях с разных вышек), то применим независимый t-test.

```python
from scipy.stats import ttest_ind


for i in tech_params[1:]:
    df_trimmed = df[df[i] < np.quantile(df[i], 0.90)]
    idx_poor = df_trimmed.Quality == 'below_avg'
    stat, p = ttest_ind(df_trimmed[idx_poor][i], df_trimmed[~idx_poor][i])
    print(f"{i:42} p-value={p:.4f}")
```

Вывод: для всех технических показателей, кроме Uplink Throughput(Kbps), отвергается нулевая гипотеза о одинаковости распределений, следовательно принимается альтернативная гипотезе: с уровнем достоверности 5% пользователи поставили оценку ниже среднего из-за более худшего качества связи. Причина провала отмены H0 для Uplink Throughput(Kbps) является черезчур агресивная обрезка выбросов (т.е. отрезали достаточно много тех данных, где добольные абоненты отправляли данные с высокой скоростью). Так как распределения не являлись нормальными (хоть и приближались к нему после обрезки), доверия t-test'у не очень много. Следовательно, проверим бутстрэпингом.

#### 2. Bootstrapping

Проверим нулевую гипотезу методом bootstrapping.

```python
def bootstrap(arr, n=100, n_trial=1000, func=np.mean):
    """
    Subsample arr and calulate func metric on it on each trial.
    Returns array of metric values of size n_trials
    """
    res = []
    n = min(len(arr),n)
    # Инициализируем цикл попыток
    for _ in range(n_trial):
        # Делаем подвыборку размера n
        if len(arr.shape) > 1:
            ind = np.random.choice(arr.shape[0], size = (n,))
            subsample = arr.iloc[ind]
        else:
            subsample = np.random.choice(arr, size = (n,))
        # рассчитываем статистику
        stat = func(subsample)
        res.append(stat)
    return res # shape (N_TRIAL,)


def run_bootstrap(data, quantile, condition, params, n_samples=1500, n_trials=4000,
                  title="Bootstrapping distributions", centered=False):
    """
    Do bootstrapping of data dataset by all tech_params and Q1_answers plotting results in 8*2 grid.
    Inputs:
        data - DataFrame containing params keys of columns
        quantile - quantile value for trimming a channel from right side
        condition - a string expression which is valid for DataFrame.query or eval functions,
            for example "Quality=='below_avg'", thus allows to separate 2 groups in data
        params - a list of keys in data to be tested
        n_samples - number of samples for subsampling, actual value is 
            min(n_samples, len(group)), where group is either data[mask] or data[~mask]
        n_trials - number of time subsampling is done
        title - general title of the plot
        centered - wether bootstrapping distributions are centered or not
    Return:
        dict of p-values for corresponing parameter
    """
    plt.figure(figsize=(15, 15))
    
    for i, param in enumerate(params):
        # no trimming for ratings
        if param == 'Q1_answer':
            df_trimmed = data
        else:
            df_trimmed = data[data[param] < np.quantile(data[param], quantile)]
        
        # mask for a group defined by condition, the second group is defined by ~mask
        mask = df_trimmed.eval(condition)
        bs_A = bootstrap(df_trimmed[mask][param], n=n_samples, n_trial=n_trials)
        bs_B = bootstrap(df_trimmed[~mask][param], n=n_samples, n_trial=n_trials)
        diff = np.array(bs_B) - np.array(bs_A)
        mean_value = np.abs(df_trimmed[~mask][param].mean() - df_trimmed[mask][param].mean())
        
        if centered:
            diff = diff - diff.mean()
            edge_r = mean_value
            edge_l = -mean_value
        else:
            edge_r = diff.mean() + mean_value
            edge_l = diff.mean() - mean_value
        
        p_value =(sum(diff <= edge_l) + sum(diff >= edge_r)) / n_trials
        
        ax = plt.subplot(4, 2, i+1)
        sns.histplot(diff, element='step', kde=True, ax=ax)
        ax.axvline(edge_r, color='g')
        ax.axvline(edge_l, color='g')
        ax.set_title(f'{param}, p-value={p_value:.4f}')

    plt.suptitle(title, y=1)
    plt.tight_layout(pad=2.0)
    plt.show()
```

```python
title = "Бутстраппинг распределение разности выборочных средних для технических\n" +\
        "показателей между группами с плохой оценкой качества связи (1-5) и хорошей(6-10)"
n_samples = 1500
n_trials = 4000

run_bootstrap(
    data=df, 
    quantile=0.9, 
    condition="Quality=='below_avg'",
    params=tech_params[1:],
    n_samples=n_samples, 
    n_trials=n_trials, 
    title=title, 
    centered=False)
```

### 4. Гипотеза о жалобе 1 и 2.

**H1**: Технические показатели этих групп абонентов имеют разное распределение, а следовательно причины обрывов  и долгого ожидания звонков различимы.

Из анализа исключим тех, кто ответил одновременно на Q2_1 и Q2_2, чтобы не вмешивать их ни в одну, ни в другую группы.

```python
df.query('Q2_1==1 and not (Q2_2==1)').describe()
```

```python
df.query('Q2_2==1 and not (Q2_1==1)').describe()
```

Количество абонентов в этих группах отличается более чем в 3 раза. Поэтому сравнивать средние значения напрямую было бы не правильно, а следовательно - всё тот же статистический тест.

#### 1. EDA

Гистограмма в этом случае не самый подходящий спобос отображения распределений с целью ивидеть разницу, так как размеры подвыборок существенно отличаются. Следовательно, смотрим только на CDP, а гистограммы оставим для оценки количества выбросов.

```python
# generate distribution plots for technical parameters
title = 'Distribution analysis of parameters,' +\
        ' compain Q2_1 (недозвоны и обрывы) vs Q2_2 (время ожидания гудков)'
distribution_alalysis(
    data=df.query("(Q2_1==1 or Q2_2==1) and not (Q2_1==1 and Q2_2==1) "),
    params=tech_params[:-1],
    condition="Q2_1==1",
    category_names=['недозвоны', 'ожидание гудуов'],
    title=title,
)
```

Гистограммы не репрезентативные, но CDP не зависит от размера группы, следовательно распределение сравнивать можно. Интерсно, что все параметры имеют одинаковое распределение, за исключением пинг web страниц, который различается более существенно. В целом группы похожи.

#### 4. Статистический тест

```python
moments = dict()
df_strata = df.query("(Q2_1==1 or Q2_2==1) and not (Q2_1==1 and Q2_2==1) ")
for param in tech_params:
    data = df_strata[df_strata[param] < np.quantile(df_strata[param], 0.97)]
    moments[param] = {
        'skewness': skew(data[param]),
        'kurtosis': kurtosis(data[param], fisher=True)
    }
pd.DataFrame(moments).transpose()
```

```python
df_strata = df.query("(Q2_1==1 or Q2_2==1) and not (Q2_1==1 and Q2_2==1) ")
for param in tech_params[1:]:
    if param == 'Q1_answer':
        df_trimmed = df_strata
    else:
        df_trimmed = df_strata[df_strata[param] < np.quantile(df_strata[param], 0.97)]
    idx = df_trimmed.Q2_1 == 1
    
    stat, p = ttest_ind(df_trimmed[idx][param], df_trimmed[~idx][param])
    print(f"{param:42} p-value={p:.4f}")
```

Видим, что p-value подтверждает нулевую гипотезу, кроме Web Average TCP RTT(ms). Данным t-test'а доверия еще меньше, так как выборки меньше, а skewness и kurtosis сильнее отличаются от нормального распределения (или в данном случае правильнее сказать t-распределения, вследствие малого размера выборки).

Для надежности проверяем гипотезу bootstrapping'ом.

#### 2. Bootstrapping

```python
title = "Бутстраппинг распределение разности выборочных средних для технических\n" +\
        "показателей между группами с жалобами Q2_1 против Q2_2"
n_samples = 304
n_trials = 10000

run_bootstrap(
    data=df.query("(Q2_1==1 or Q2_2==1) and not (Q2_1==1 and Q2_2==1)"), 
    quantile=0.97, 
    condition="Q2_1==1",
    params=tech_params[1:],
    n_samples=n_samples, 
    n_trials=n_trials, 
    title=title
)
```
